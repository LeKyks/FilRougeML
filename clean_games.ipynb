```python
import pandas as pd

# Charger le fichier CSV dans un DataFrame
file_path = 'C:/Users/Salma/OneDrive/Documents/Deep learning/FILL-ROUGE/steam-200k.csv'
df = pd.read_csv(file_path, header=None, names=['user_id', 'game', 'behavior', 'hours', '0'])

# Afficher les premières lignes du DataFrame pour vérifier
df.head()
```

```python
# Supprimer la colonne '0' du DataFrame
df = df.drop('0', axis=1)

# Afficher les premières lignes du DataFrame mis à jour
df.head()
```

```python
# Identifier les jeux qui ont été achetés mais non joués
game_purchased_no_played = df[df['behavior'] == 'purchase']

# Filtrer les jeux qui ont également l'action "play"
game_purchased_no_played = game_purchased_no_played[~game_purchased_no_played['game'].isin(df[df['behavior'] == 'play']['game'])]

# Afficher les jeux achetés mais non joués
game_purchased_no_played.head()
```

```python
# Sauvegarder les résultats dans un nouveau fichier CSV
result_file_path = 'C:/Users/Salma/OneDrive/Documents/Deep learning/FILL-ROUGE/game_purchased.csv'
game_purchased_no_played.to_csv(result_file_path, index=False)
```

```python
# Charger la première table (jeux achetés)
path_achetes = 'C:/Users/Salma/OneDrive/Documents/Deep learning/FILL-ROUGE/game_purchased.csv'
df_achetes = pd.read_csv(path_achetes)

# Charger la deuxième table (jeux joués)
path_joues = 'C:/Users/Salma/OneDrive/Documents/Deep learning/FILL-ROUGE/steam-200k_clean.csv'  # Remplacez par le chemin de votre fichier
df_joues = pd.read_csv(path_joues)
```

```python
# Concaténer les deux DataFrames
concatenated_df = pd.concat([df_achetes, df_joues], ignore_index=True)
concatenated_df.head()
```

```python
# Sauvegarder le DataFrame concaténé en fichier CSV
result_file_path = 'C:/Users/Salma/OneDrive/Documents/Deep learning/FILL-ROUGE/game_played_purchesed.csv'
concatenated_df.to_csv(result_file_path, index=False)
```
